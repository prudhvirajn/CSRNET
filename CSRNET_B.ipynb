```
tf.keras.backend.clear_session()
```

```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```
img_inputs = keras.Input(shape=(None,None,3), name='img')

x = layers.Conv2D(64, 3, activation='relu', padding='same')(img_inputs)
x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
x = layers.MaxPooling2D(strides=2)(x)

x = layers.Conv2D(128, 3, activation='relu', padding='same')(x)
x = layers.Conv2D(128, 3, activation='relu', padding='same')(x)
x = layers.MaxPooling2D(strides=2)(x)

x = layers.Conv2D(256, 3, activation='relu', padding='same')(x)
x = layers.Conv2D(256, 3, activation='relu', padding='same')(x)
x = layers.Conv2D(256, 3, activation='relu', padding='same')(x)
x = layers.MaxPooling2D(strides=2)(x)

x = layers.Conv2D(512, 3, activation='relu', padding='same')(x)
x = layers.Conv2D(512, 3, activation='relu', padding='same')(x)
x = layers.Conv2D(512, 3, activation='relu', padding='same')(x)

x = layers.Conv2D(512, 3, activation='relu', padding='same', dilation_rate=2)(x)
x = layers.Conv2D(512, 3, activation='relu', padding='same', dilation_rate=2)(x)
x = layers.Conv2D(512, 3, activation='relu', padding='same', dilation_rate=2)(x)

x = layers.Conv2D(256, 3, activation='relu', padding='same', dilation_rate=2)(x)
x = layers.Conv2D(128, 3, activation='relu', padding='same', dilation_rate=2)(x)
x = layers.Conv2D(64, 3, activation='relu', padding='same', dilation_rate=2)(x)
output = layers.Conv2D(1, 1, activation='relu', padding='same', name='output')(x)
```

```
model = keras.Model(inputs=img_inputs, outputs=output, name='csrnet_b')
```

```
model.summary()
```
